```python
def percentile(data, percentile):
    size = len(data)
    # get the index of the percentile element
    y = int(math.ceil((size * percentile) / 100)) - 1
    # sort the data
    x = sorted(data)
    # return data at percentile index
    return x[y]
```

```python
pl_95_data1 = percentile(xl_data3['NIS'], 95)
pr_95_data1 = percentile(xr_data3['NIS'], 95)
pl_95_data2 = percentile(xl_data4['NIS'], 95)
pr_95_data2 = percentile(xr_data4['NIS'], 95)
```

```python
fig = plt.figure(figsize=(20, 5))
fig.suptitle('NIS', fontsize=20, fontweight='bold')

ax = fig.add_subplot(121)
fig.subplots_adjust(top=0.85)
ax.set_title('Dataset 1 Laser', fontsize=18, fontweight='bold')
ax.set_xlabel('time')
ax.set_ylabel('chi2')
ax.plot(xl_data3['NIS'])
ax.plot((0, 500), (pl_95_data1, pl_95_data1), 'k-', linewidth=3, label = 'p95(2 std)')
ax.legend(loc='upper left')
ax.axis([0, 500 , 0, 35])

ax = fig.add_subplot(122)
fig.subplots_adjust(top=0.85)
ax.set_title('Dataset 1 Radar', fontsize=18, fontweight='bold')
ax.set_xlabel('time')
ax.set_ylabel('chi2')
ax.plot(xr_data3['NIS'])
ax.plot((0, 500), (pr_95_data1, pr_95_data1), 'k-', linewidth=3, label = 'p95(2 std)')
ax.legend(loc='upper left')
ax.axis([0, 500 , 0, 35])

plt.show()
```
